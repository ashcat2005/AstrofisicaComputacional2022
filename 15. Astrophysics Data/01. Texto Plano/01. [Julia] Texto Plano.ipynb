## 1. Lectura de archivos de texto plano

Utilizando [DelimitedFiles](https://docs.julialang.org/en/v1/stdlib/DelimitedFiles/), es posible leer desde un archivo de texto plano y cargarlos directamente a un arreglo. 

```julia
using DelimitedFiles
using Plots

data = readdlm("example01.txt", skipstart=1)
data
```

```julia
plot(data[:,1],data[:,2])
```

Cuando los datos en el archivo están separados por un simbolo diferente al espacio, debe escribirse el argumento corrspondiente

```julia
data = readdlm("example02.txt", skipstart=1)
```

```julia
data = readdlm("example02.txt", ',', skipstart=1)
```

```julia
plot(data[:,1],data[:,2])
```

## 02. Escritura de archivos de texto plano

De nuevo, la librería [DelimitedFiles](https://docs.julialang.org/en/v1/stdlib/DelimitedFiles/) tiene una función sencilla para la escritura de archivos de texto plano.

```julia
function f(x)
    return exp(-x^2/5)
end

wdata = zeros(100,2)

wdata[:,1] = LinRange(-10,10,100)
wdata[:,2] = f.(wdata[:,1])


writedlm("out_file1[julia].txt", wdata)
```

```julia
# Para incluir un header
open("out_file2[julia].txt", write=true) do f
    write(f, "# This is the header of the file\n")
    writedlm(f, wdata)
end
```

```julia
data = readdlm("out_file2[julia].txt", skipstart=1)
plot(data[:,1],data[:,2])
```

```julia
open("out_file3[julia].txt", write=true) do f
    write(f, "# This is the header of the file\n")
    writedlm(f, wdata, ',')
end


data = readdlm("out_file3[julia].txt", ',', skipstart=1)
plot(data[:,1],data[:,2])
```
